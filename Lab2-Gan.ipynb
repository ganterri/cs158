```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

## Problem 1

```python
# FOR UNIT OF TIME BEING A DAY

# Given N, d, k, b
N = 3_000_000
d = 3
k = 1
b = 1

# define the r, a parameters
r = (k*b)/N
a = 1/d

# define the t0 and tf parameters
t0 = 0
tf = 50

#define the initial conditions
I_0 = 5
S_0 = N - I_0
R_0 = 0
y0 = np.array([S_0, I_0, R_0])

# define the ode system as given in the SIR problem
def sir(t,y):
    S = y[0]
    I = y[1]
    R = y[2]
    
    dS_dt = -r*S*I
    dI_dt = r*S*I - a*I
    dR_dt = a*I
    
    return np.array([dS_dt, dI_dt, dR_dt])

# solve the system
sol = solve_ivp(sir, (t0,tf), y0, t_eval=np.arange(t0, tf+0.1, 0.5))

# plot t against y[0], y[1], and y[2]
plt.plot(sol.t, sol.y[0]/N, 'black', label='Susceptible')
plt.plot(sol.t, sol.y[1]/N, 'r', label='Infected')
plt.plot(sol.t, sol.y[2]/N, 'blue', label='Recovered')
plt.xlabel('T (days)')
plt.ylabel('Proportion of Population')
plt.title('With Unit Time as Per Day')
plt.legend()
plt.show()
```

```python
#np.where(sol.y[1]==sol.y[1].max())
#(array([42], dtype=int64),)

i = 42
date = sol.t[i]
susceptibles = sol.y[0][i]
susceptibles_percent = (sol.y[0][i]/N)*100
recovereds = sol.y[2][i]
recovereds_percent = (sol.y[2][i]/N)*100

print(date)
print(susceptibles)
print(susceptibles_percent)
print(recovereds)
print(recovereds_percent)
```

## Problem 3


```python
# Given N, d, k, b, f, mu
N = 7_000_000_000
d = 10
k = 1/2
b = 1
f = 1/50
mu = 0.0001

# define the r, a parameters
a = 1/d
r = (k*b)/N

# define the t0 and tf parameters
t0 = 0
tf = 365

#define the initial conditions
S_0 = N-1000
I_0 = 1000
R_0 = 0
y0 = np.array([S_0, I_0, R_0])

# define the ode system as given in the SIR problem
def sirs(t,y):
    S = y[0]
    I = y[1]
    R = y[2]
    
    dS_dt = f*R + mu*(1 - S) - r*S*I
    dI_dt = r*S*I - (a + mu)*I
    dR_dt = -f*R + a*I - mu*R
    
    return np.array([dS_dt, dI_dt, dR_dt])

# solve the system
sol = solve_ivp(sirs, (t0,tf), y0, t_eval=np.arange(t0, tf+.1, 0.5))

# plot t against y[0], y[1], and y[2]
plt.plot(sol.t, sol.y[0]/N, label='Susceptible')
plt.plot(sol.t, sol.y[1]/N, label='Infected')
plt.plot(sol.t, sol.y[2]/N, label='Recovered')
plt.xlabel('Day')
plt.ylabel('Proportion of Population')
plt.legend()
plt.show()
```
